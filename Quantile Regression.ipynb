```python
import pandas as pd
df = pd.read_csv('allteams-nice-NI.csv')
df
```

```python
df_nona = df.dropna()
prune = df_nona[(df_nona['blitz-rd']<100)&(df_nona['puzzle-rd']<100)&
                        (df_nona['rapid-rd']<100)&(df_nona['storm-runs']>10)]
prune
```

```python
qprune = prune[['puzzle-rating','rapid-rating','blitz-rating','storm-score']]
qprune.columns = ['puzzle','rapid','blitz','storm']
qprune
```

```python
import statsmodels.formula.api as smf
qmod = smf.quantreg('blitz ~ storm + puzzle', qprune)
res = qmod.fit(q=.5)
print(res.summary())
```

```python
import numpy as np
quantiles = np.arange(0.1,1,0.1)
print(quantiles)
models = []
params = []
```

```python
for qt in quantiles:
    print(qt)
    res = qmod.fit(q = qt )
    models.append(res)
    params.append([qt, res.params['Intercept'], res.params['storm'],res.params['puzzle']] 
                  + res.conf_int().loc['storm'].tolist() + res.conf_int().loc['puzzle'].tolist())
```

```python
params = pd.DataFrame(data = params, columns = ['qt','intercept','storm_coef','puzzle_coef'
                                                ,'scf_lower_bound','scf_upper_bound'
                                               ,'pcf_lower_bound','pcf_upper_bound'])

params
```

```python
params.to_csv('BlitzQuantReg-10r-planes.csv')
```
