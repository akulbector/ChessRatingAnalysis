```python
import pandas as pd
import json
import ast
```

```python
df = pd.read_csv('allteams.csv')
```

```python
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
table = []
```

```python
df['perfs'][0]
```

```python
for i in range(len(df)):
    row = []
    row.append(df['id'][i])
    perfs = ast.literal_eval(df['perfs'][i])
    if 'puzzle' in perfs.keys():
        row.append(perfs['puzzle']['games'])
        row.append(perfs['puzzle']['rating'])
        row.append(perfs['puzzle']['rd'])
    else:
        row.append(None)
        row.append(None)
        row.append(None)
    if 'blitz' in perfs.keys():
        row.append(perfs['blitz']['games'])
        row.append(perfs['blitz']['rating'])
        row.append(perfs['blitz']['rd'])
    else:
        row.append(None)
        row.append(None)
        row.append(None)
    if 'rapid' in perfs.keys():
        row.append(perfs['rapid']['games'])
        row.append(perfs['rapid']['rating'])
        row.append(perfs['rapid']['rd'])
    else:
        row.append(None)
        row.append(None)
        row.append(None)
    if 'classical' in perfs.keys():
        row.append(perfs['classical']['games'])
        row.append(perfs['classical']['rating'])
        row.append(perfs['classical']['rd'])
    else:
        row.append(None)
        row.append(None)
        row.append(None)
    if 'storm' in perfs.keys():
        row.append(perfs['storm']['runs'])
        row.append(perfs['storm']['score'])
    else:
        row.append(None)
        row.append(None)
    if type(df['profile'][i])!=float:
        if 'fideRating' in ast.literal_eval(df['profile'][i]).keys():
            row.append((ast.literal_eval(df['profile'][i]))['fideRating'])
        else:
            row.append(None)
    else:
        row.append(None)
    table.append(row)
df2 = pd.DataFrame(table,columns=['id','puzzle-games','puzzle-rating','puzzle-rd','blitz-games','blitz-rating','blitz-rd',
                                     'rapid-games','rapid-rating','rapid-rd','classical-games','classical-rating',
                                  'classical-rd','storm-runs','storm-score','fide'])
df2
```

```python
df2.to_csv('allteams-nice-wfide.csv',index=False)
```
